## 1. Setup and Imports

```python
# Jupyter notebook setup
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import logging

# Core QICK imports
from qick import *

# tProc v2 specific classes
from qick.asm_v2 import AveragerProgramV2, QickSweep1D, QickSpan

# Optional: increase logging level for debugging
# logging.basicConfig(level=logging.DEBUG)
logging.basicConfig(level=logging.INFO, format='%(levelname)-8s [%(filename)s:%(lineno)d] %(message)s')
# Reduce noise from the tProc core
logging.getLogger("qick_processor").setLevel(logging.WARNING)

# Display the loaded QICK library version
print(f"QICK Version: {qick.get_version()}")
```

**Explanation:** This cell imports all necessary libraries. The `qick.asm_v2` module contains the classes for building tProc v2 programs. The logging configuration helps with debugging if something goes wrong.

## 2. Connect to the RFSoC Board

You need to provide the correct path to your firmware `.bit` file. This file is specific to your board model and the firmware version.

```python
# --- IMPORTANT: Change this path to point to your firmware file ---
# Common paths for different boards (examples):
# ZCU111:   '/home/xilinx/jupyter_notebooks/qick_firmware/out/qick_zcu111.bit'
# RFSoC4x2: '/home/xilinx/jupyter_notebooks/qick_firmware/out/qick_4x2.bit'
# ZCU216:   '/home/xilinx/jupyter_notebooks/qick_firmware/out/qick_216.bit'

BITSTREAM_PATH = '/path/to/your/firmware.bit'  # <--- YOU MUST CHANGE THIS

# Connect to the board
soc = QickSoc(BITSTREAM_PATH)
soccfg = soc  # 'soccfg' is a common alias used throughout QICK examples

print("\n=== Board Connection Successful ===")
```

## 3. Inspect the Hardware Configuration

The `QickSoc` object contains all the information about your firmware: clock frequencies, generator channels, readout channels, digital I/O, and tProc capabilities.

```python
print(soccfg)
```

**Explanation:** The printed output shows:
- **Global clocks:** tProc timing clock, RF reference frequency
- **Signal generators:** Number of channels, sample rates, envelope memory size
- **Readout channels:** Sample rates, buffer sizes, trigger mappings
- **Digital output pins:** Available PMOD pins
- **tProc configuration:** Program memory size, data memory size, waveform memory size, core clock

This information is essential when configuring your experiments.

## 4. Quick Hardware Test

To verify that the board is responding correctly, read the tProc's program counter and test data memory access.

```python
# Read the tProc's program counter (should be 0 after reset)
pc = soc.get_tproc_counter(addr=1)
print(f"tProc Program Counter: {pc}")

# Read a data memory address to verify communication
# Data memory address 0 should contain some value, but reading it confirms the interface works
try:
    data = soc.read_mem(length=1, mem_sel='dmem', addr=0)
    print(f"Data memory read test: {data}")
except Exception as e:
    print(f"Data memory read failed: {e}")

print("\n✓ Basic board communication verified via program counter read.")
```

**Explanation:** These simple reads confirm that the communication with the board is working correctly. The tProc program counter should be 0 before we load a program.
